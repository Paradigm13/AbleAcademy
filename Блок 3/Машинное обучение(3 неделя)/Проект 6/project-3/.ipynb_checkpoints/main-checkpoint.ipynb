Следующим шагом нужно закодировать категориальные данные. Так как у нас три столбца с категориальными значениями, каждая из которых имеет от 2 до 4 уникальных значений, можно воспользоваться классом LabelEncoder из библиотеки sklearn.preprocessing

````python
from sklearn.preprocessing import LabelEncoder

def label_encoding(data):
    
    columns = data.select_dtypes(include = 'object').columns
    encode = LabelEncoder()
    
    for col in columns:
        data[col] = encode.fit_transform(data[col])
        
    return data
````

````python
train_data = label_encoding(train_data)
train_data
````

````python
test_data = label_encoding(test_data)
test_data
````

# 2. Обучение и проверка моделей.

Для обучения моделей будем использовать LinearRegression, RandomForestRegressor и PolynomialRegression.

````python
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.preprocessing import PolynomialFeatures
import sklearn.metrics as metrics
````

LinearRegression

````python
def lin_regressor(data, target):
    regressors = {}
    
    X = data.drop([target], axis = 1)
    y = data[target]

    lin_regressor = LinearRegression()
    lin_regressor.fit(X, y)
    
    regressors['regressor'] = lin_regressor
    regressors['parameters'] = X.columns
    regressors['features'] = ''

    return regressors
````

````python
def validation(data, model, target):
    y_true = data[target].values
    
    if isinstance(model['features'], PolynomialFeatures):
        pol_degree = model['features']
        X = pol_degree.fit_transform(data[model['parameters']])
        y_pred = model['regressor'].predict(X)
    else:
        y_pred = model['regressor'].predict(data[model['parameters']])
    
    result = pd.DataFrame({target: y_true.flatten()})
    result['-'.join(model['parameters'])] = y_pred
    
    rmse = round(metrics.mean_squared_error(y_true, y_pred, squared = False), 4)
    mae = round(metrics.mean_absolute_error(y_true, y_pred), 4)
    R2 = round(metrics.r2_score(y_true, y_pred), 4)

    all_metrics = pd.DataFrame({'RMSE': rmse,
                                'MAE': mae,
                                'R2': R2}, index=[0])
    return result, all_metrics
````

````python
lin_model = lin_regressor(data = train_data, target = 'charges')
lin_predict, lin_metrics = validation(data = test_data, model = lin_model, target = 'charges')
````

````python
lin_predict.head()
````

````python
lin_metrics
````

Для модели линейной регрессии, R2, равный 78 процентам, считается довольно хорошим результатом.

RandomForestRegressor

````python
def random_forest_regressor(data, target, n_estimators = None, random_state = None, max_depth = None):
    regressors = {}
    
    X = data.drop([target], axis = 1)
    y = data[target]
    
    regressor = RandomForestRegressor(n_estimators = n_estimators, random_state = random_state, max_depth = max_depth)
    regressor.fit(X, y)
    
    regressors['regressor'] = regressor
    regressors['parameters'] = X.columns
    regressors['features'] = ''
    
    return regressors
````

````python
rfr_model = random_forest_regressor(data = train_data, target = 'charges', n_estimators = 130, random_state = 1, max_depth = 5)
rfr_predict, rfr_metrics = validation(data = test_data, model = rfr_model, target = 'charges')
````

````python
rfr_predict.head()
````

````python
rfr_metrics
````

Модель случайного леса предсказывает на порядок выше, нежели линейная регрессия, коэффициент детерминации тому подтверждение.

PolynomialRegression

````python
def pol_regression(data, target, degree, interaction_only = None):
    
    regressors = {}

    X = data.drop([target], axis = 1)
    y = data[target]
    
    pol_feature = PolynomialFeatures(degree = degree, interaction_only = interaction_only)
    pol_degree_feature = pol_feature.fit_transform(X.values)
    pol_degree_regressor = LinearRegression().fit(pol_degree_feature, y.values)

    regressors['regressor'] = pol_degree_regressor
    regressors['parameters'] = X.columns
    regressors['features'] = pol_feature
    
    return regressors
````

````python
polreg_model = pol_regression(data = train_data, target = 'charges', degree = 2, interaction_only = False)
polreg_predict, polreg_metrics = validation(data = test_data, model = polreg_model, target = 'charges')
````

````python
polreg_predict
````

````python
polreg_metrics
````

Модель обученная через PolynomialFeature показывает результат хуже, чем при RandomForestRegressor

# 3. Сравнение моделей на основе метрик регрессии.

````python
polreg_metrics = polreg_metrics.rename(index = {0: "PolynomialFeature"})
rfr_metrics = rfr_metrics.rename(index = {0: "RandomForestRegressor"})
lin_metrics = lin_metrics.rename(index = {0: "LinearRegressor"})

all_metrics = pd.concat([polreg_metrics, rfr_metrics, lin_metrics], ignore_index = False).rename_axis('Model')
````

````python
all_metrics.sort_values(by = 'R2', ascending = False)
````

````python
fig, axes = plt.subplots(nrows=3, ncols=1, figsize=(8, 18))
colors = plt.cm.Set2(range(len(all_metrics.columns)))
for idx, metr in enumerate(all_metrics.columns):
    axes[idx].bar(all_metrics.index, all_metrics[metr], color = colors)
    axes[idx].set_ylabel(f'{metr}')
    axes[idx].set_title(f'{metr} для разных моделей')
````

По полученным результатам, самая лучшая модель - RandomForestRegressor
