````python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
%matplotlib inline
````

````python
rates = pd.read_csv('./foreign_exchange_rates.csv')
rates.drop('Unnamed: 0', axis = 1, inplace = True)
rates.head()
````

````python
'''
Для начала, нужно разделить дату на год, месяц и год-месяц.
Это нужно для того, чтобы посмотреть, какие были изменения в курсе валют, с течением времени.
'''

rates["Time Serie"] = pd.to_datetime(rates['Time Serie'])

rates['Year'] = rates['Time Serie'].dt.year
rates['Month'] = rates['Time Serie'].dt.month
rates['Month_year'] = rates['Time Serie'].dt.to_period('M')
````

````python
rates.dtypes
````

````python
'''
Как можно заметить, типы данных всех валютных пар у нас в виде object, а для того, чтобы сделать какие-нибудь анализы, нужно 
перевести эти данные во float. Также, у нас в dataframe есть значения "ND", которые будут заменены на NaN, и в последующем,
методом interpolate(), будут заменены средним значением двух соседних чисел.
'''
rates_copy = rates.copy()
rates_copy = rates_copy.replace('ND', np.nan)

object_cols = rates_copy.iloc[:, :-1].select_dtypes(include=['object']).columns.tolist()
rates_copy[object_cols] = rates_copy[object_cols].astype(float).interpolate()
````

````python
# Корреляция только по валютным парам
corr_matrix = rates_copy.corr().iloc[:-2,:-2]
plt.figure(figsize=(16, 16))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm');
````

````python
'''
•По матрице корреляции видно, что в основном сильные валюты коррелируют друг с другом больше всего.
•Страны, у которых валюты используется в узком кругу, практически не коррелируются друг с другом. 
•В пример возьмём валютные пары AUD/USD и EUR/USD. Одной из причин высокой корреляции с другими сильными валютами является то,
 что Австралия занимает второе место в мировом рейтинге стран-производителей золота. Таким образом, этот товар имеет большое 
 влияние на экономику страны, и, следовательно, пара AUD/USD коррелирует с динамикой цены на золото.
•Пара EUR/USD является мажорной валютой, и высоко коррелируется относительно других сильных валют линейно.
•Высокая корреляция наблюдается у пар YUAN/USA с SGD/USA. Связано это с тем, что Китай и Сингапур являются крупными
 экономическими партнерами, особенно в торговле. Китай является крупнейшим торговым партнером Сингапура, а Сингапур - важным 
 центром для китайских инвесторов и компаний, которые ищут доступ к международным рынкам.
•Корреляция равна 1 между датской кроной и евро, что означает, что две валюты полностью синхронизированы и движутся в одном 
 направлении. Это происходит из-за фиксированного обменного курса между датской кроной и евро. Дания является членом 
 Европейского союза, но не присоединилась к еврозоне и сохраняет свою национальную валюту - датскую крону. Однако, чтобы 
 облегчить торговлю и инвестиции, Дания установила фиксированный обменный курс своей валюты с евро. Это означает, что цена 
 датской кроны относительно евро остается постоянной, и валюты движутся в одном направлении.
•Шведская крона имеет такой же случай, как с датской кроной, по отношению к евро. Швеция и еврозона имеют тесные экономические 
 связи, в том числе в торговле, инвестициях и финансовых потоках. Кроме того, Швеция и еврозона сильно зависят от мировых 
 экономических и финансовых факторов, таких как изменения процентных ставок, рост экономики и т.д.
•Новозеландский доллар и австралийский доллар имеют высокую корреляцию. Это связано с несколькими факторами, включая близость 
 географического положения, сходство экономических структур и торговых партнеров, а также сопряженность их денежно-кредитных 
 систем. Новая Зеландия и Австралия являются соседними странами, имеющими общие экономические интересы и взаимодействия.
 Новая Зеландия и Австралия обе имеют доллары как национальные валюты, и обе центральные банки имеют сходные политики и 
 практики в области денежно-кредитной политики.
•Канадский и австралийский доллары. оба доллара относятся к группе комодити-валют, что означает, что они часто реагируют на 
 изменения в ценах на сырьевые товары на мировых рынках. Это может создавать сходства в движениях курсов между канадским и 
 австралийским долларами. Таким образом, корреляция между канадским долларом и австралийским долларом весьма высокая.
•Корреляция между швейцарским франком и новозеландским долларом по матрице корреляции высокая, но она не должна быть таковой.
 Швейцария находится в Европе, а Новая Зеландия - в Тихоокеанском регионе, это может приводить к различиям в экономической 
 ситуации и изменениям валютного курса. Тем не менее, как и многие другие валюты, швейцарский франк и новозеландский доллар 
 могут быть подвержены влиянию мировых экономических событий, таких как изменения в процентных ставках и глобальных ценах на 
 сырьевые товары, что может привести к короткосрочным сходствам в движениях курсов между двумя валютами. 

''';
````

````python
columns = rates_copy.columns[1:23]

grouped_year = rates_copy.groupby('Year')[columns].mean()

fig = plt.figure(figsize = (20,20))
for i, col in enumerate(grouped_year):
        axs = fig.add_subplot(6, 4, i + 1)
        axs.plot(rates_copy["Year"].unique(), grouped_year[col], '-o', mfc = 'r')
        axs.set_title(col, fontsize = 15)
        axs.set_xticks(rates_copy["Year"].unique())
        axs.set_xlabel("Year")
        axs.set_ylabel("Exchange rate")
        plt.setp(axs.get_xticklabels(), rotation=45)

plt.tight_layout()
````

````python
'''
Полученные графики показывают средние значения каждой валютной пары в определенные года. По этим графикам можно посмотреть, как
курс той или иной валютной пары менялся с течением времени.
Во многих графиках видно, что в 2007 - 2008 годах, из-за мирового кризиса, валюты по отношению к доллару сильно обесценивались. 
В пример возьму GBP/USD (United Kingdom - UK Pound/US$). В 2007 - 2008 годах произошло резкое обесценивание фунта стерлингов, 
поскольку Великобритания очень сильно пострадала от кредитного кризиса, что и показывает график. 
Валютные пары EUR/USD и DKK/USD и их курсы менюятся идентично с истечением времени, что говорит о высокой корреляции.
Еще можно заметить, что курсы мажоритарных валют двигаются схоже по времени.
Шри-Ланская рупия по отношению к доллару с каждым годом обесценивается. Так, в начале 2000-х годов можно было за один долар 
взять чуть меньше 80 рупий, а в 2019 году, за этот же доллар можно взять 180 рупий. Связано это из=за высокой инфляции в 
Шри-Ланке.
У индийского рупия также наблюдается сильное обесценивание. С 2000 по 2011 год, котировка валютной пары INR/USD находилась в 
диапазоне от 40 до 50, и начиная с 2011 года начала стремительно подниматься.
Швейцарские франки, исходя из графика, за последние 5 лет находятся практически на одном уровне, без сильных скачков. 
Исходя из диапазона курсов, самыми стабильными являются валютные пары EUR/USD, GBP/USD, HKD/USD
Как видно на графике, курс японской йены во время кризиса сильно проседала, и после вернулась к своему начальному курсу.

''';
````

````python
groupby_top_countries = rates_copy.groupby('Month_year').apply(lambda x: pd.Series
           ({
              'AUD_mean': x['AUSTRALIA - AUSTRALIAN DOLLAR/US$'].mean(),
              'EUR_mean': x['EURO AREA - EURO/US$'].mean(),
              'NZD_mean': x['NEW ZEALAND - NEW ZELAND DOLLAR/US$'].mean(),
               'GBP_mean': x['UNITED KINGDOM - UNITED KINGDOM POUND/US$'].mean(),
              'SGD_mean': x['SINGAPORE - SINGAPORE DOLLAR/US$'].mean(),
              'CHF_mean': x['SWITZERLAND - FRANC/US$'].mean(),
           })).reset_index()
groupby_top_countries['USD_mean'] = [1 for i in range(240)] #для визуализации разницы всех обменных курсов по сравнению с базой.

x_top = groupby_top_countries['Month_year'].astype(str)
y_top = groupby_top_countries.iloc[:,1:]
labels = groupby_top_countries.columns[1:]
x_ticks = list(range(1, 240, 12))
x_ticklabels = [x for x in range(2000, 2020)]
plt.figure(figsize=(20,10))
ax = plt.subplot()
plt.plot(x_top, y_top)
ax.set_xticks(x_ticks)
ax.set_xticklabels(x_ticklabels)
plt.legend(labels)
plt.title("Exchange Rate: Top Countries/USD")
plt.xlabel("Year")
plt.ylabel("Exchange Rate")
plt.show()
````

````python
'''
По графику видно, что структура курсов валют по отношению к базовой валюте (прямая линия) USD, схожа. Наблюдается скачок с 
2008 по 2010 года. Это связано с Мировым кризисом.
''';
````
