````python
house_rent.query('area>1000')
````

````python
'''
По методу query() был выставлен фильтр, чтобы посмотреть, сколько всего значений, превышающих определенный предел. Учитывая, что
квадратура дома равна 1600 и арендная плата равна всего 7900 у.е., учитывая, что дом находится в городе, можно сделать вывод,
что данная строка является ошибочной. Остальные две строки, являются ошибочными, так как для таких квадратур указана этажность, 
и сумма аренды очень мала. Их мы исключим, чтобы графики не сильно искажались.
''';
````

````python
house_rent = house_rent.query('area <= 1000')

sns.boxplot(house_rent['area']);
````

````python
'''
Исключив большие значения, теперь можно наблюдать лучшее распределение.

Далее нам нужно будет подготовить строки для построения графиков. 
Для начала нужно будет заменить прочерк(-) на значение 0, что будет означать, что это дома.
Затем надо обработать столбцы со значениями аренды, налогов, суммы страховки от пожаров и платы ассоциаций домовладельцев(hoa). 
''';
````

````python
house_rent_copy = house_rent.copy()
house_rent_copy = house_rent_copy.replace('-', 0)
house_rent_copy.iloc[:, 8:] = house_rent_copy.iloc[:, 8:].replace(to_replace='[^0-9]', value='', regex=True)
house_rent_copy.iloc[:, 8:] = house_rent_copy.iloc[:, 8:].replace('', 0)
house_rent_copy.iloc[:, 8:] = house_rent_copy.iloc[:, 8:].astype("int64")
house_rent_copy['floor']= house_rent_copy['floor'].astype("int64")
house_rent_copy


````

````python
'''
По boxplot можно заметить самое большое значение, которое равно 99. Прежде чем его убирать, считая за выброс, нужно проверить,
на самом ли деле, данная этажность является ошибочной. Учитывая квадрату квартиры, нахождения жилого комплекса в городе, а 
также высокую арендную плату, можно сделать вывод, что данная строка является приемлемой, и что действительно такие квартиры 
могут существовать.
''';
````

````python
fig = plt.figure(figsize = (15, 15))
    
for i, col in enumerate(house_rent_copy.iloc[:, 8:]):
    axs = fig.add_subplot(3, 2, i+1)
    axs.boxplot(house_rent_copy[col])
    axs.tick_params(axis='y', labelsize=13)

    axs.set_xlabel(col, fontsize=20)

    plt.tight_layout()
````
