````python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
%matplotlib inline
````

````python
netflix_titles = pd.read_csv('./netflix_titles.csv')
netflix_titles.head()
````

````python
'''
Так как show_id не несёт никакой информации для анализа, мы его уберём. 
''';
netflix_titles.drop('show_id', axis = 1, inplace = True)
````

````python
netflix_titles.info()
````

````python
netflix_titles.isnull().sum()
````

````python
'''
В нашем датасете имеются столбцы с пустыми значениями. Например в колонке "director" 1969 пустых строк, их мы заполним как 
"No director".
Пустые строки "cast" заполним как "No cast".
В колонке "country" заполним "Country unavailable"
Так как в колонках "date_added", "rating" мало пустых строк, то мы можем их убрать.
''';
netflix_titles['director'].fillna('No director', inplace = True)
netflix_titles['cast'].fillna('No cast', inplace = True)
netflix_titles['country'].fillna('Country anavailable', inplace = True)
netflix_titles.dropna(subset = ['date_added', 'rating'], inplace = True)
````

````python
netflix_titles.isnull().sum()
````

````python
netflix_titles.head()
````

````python
plt.figure(figsize=(12,6))

plt.title("Процентное соотношение фильмов и ТВ шоу на Netflix")

g = plt.pie(netflix_titles['type'].value_counts(), explode = (0.025,0.025), labels = netflix_titles['type'].value_counts().index, 
            colors = ['xkcd:olive', 'xkcd:sky blue'], autopct = '%1.1f%%', startangle = 180)
plt.show()
````

````python
netflix_titles['date_added'] = netflix_titles['date_added'].str.strip()
netflix_titles['date_added'] = pd.to_datetime(netflix_titles['date_added'], format='%B %d, %Y')
netflix_titles.head()
````

````python
netflix_titles.dtypes
````

````python
'''
Следующим шагом будет то, когда Netflix начал добавлять на платформу фильмы и тв-шоу. Для начала нужно будет отфильтровать 
строки относящиеся к фильмам и тв-шоу. Осью х будет служить год добавления, который будет взят из "date_added"
''';

netflix_titles = netflix_titles.assign(year_added=netflix_titles['date_added'].dt.year)
netflix_movies = netflix_titles.query('type == "Movie"').assign(year_added=netflix_titles['date_added'].dt.year)
netflix_tvshows = netflix_titles.query('type == "TV Show"').assign(year_added=netflix_titles['date_added'].dt.year)

netflix_year = netflix_titles.groupby('year_added')['type'].agg(count = 'count').reset_index()
movies_year = netflix_movies.groupby('year_added')['type'].agg(count = 'count').reset_index()
shows_year = netflix_tvshows.groupby('year_added')['type'].agg(count = 'count').reset_index()


fig, ax = plt.subplots(figsize=(13, 7))

sns.set_style("darkgrid")
sns.lineplot(data = netflix_year, x = 'year_added', y = 'count', label = 'Total')
sns.lineplot(data = movies_year, x = 'year_added', y = 'count', label = 'Movie')
sns.lineplot(data = shows_year, x = 'year_added', y = 'count', label = 'TV Show')
ax.set_xticks(np.arange(2008, 2021, 1))
plt.title('Общее количество контента, добавленного за все годы')
plt.legend()
plt.ylabel('Releases')
plt.xlabel('Year')
plt.show()
````

````python
'''
По графику можно сделать выводы, что начиная с 2014 года, началось активное добавление фильмов и ТВ шоу. С годами больше 
приоритета в добавлении фильмов нежели ТВ шоу. Резкое понижение с 2019 на 2020 год связан с тем, что данный датасет охватывает 
лишь первый месяц 2020 года.
''';
````

````python
netflix_titles['date_added'][netflix_titles['date_added'] >= '2020'].value_counts()
````

````python
'''
Следующее наблюдение которое хотелось бы затронуть, это страны, выпускащие больше всего фильмов. У нас имеются строки, в которых
прописаны несколько стран через запятую, для начала их нужно разделить, и затем уже строить графики.
Методом query отфильтруем неизвестные страны.
''';
netflix_countries = netflix_titles.query('country != "Country anavailable"').set_index('title')['country'].str.split(',', expand = True).stack().reset_index(level = 1, drop = True)

plt.figure(figsize = (14,7))
ax = sns.countplot(y = netflix_countries, order = netflix_countries.value_counts().index[:20])
plt.title('Топ 20 стран по выпуску контента на Netflix')
plt.xlabel('Titles')
plt.ylabel('Country')
plt.show()
````

````python
'''
По полученному графику видно, что больше всего производимого контента приходится на США. Самые крупные киностудии приходятся 
именно на эту страну. 
''';
````

````python
'''
Также, можно сделать анализ по режиссерам, которые отсняли больше всего фильмов. В данной колонке имеются строки, в которых
прописаны несколько режиссеров, их также нужно будет разделить.
''';
netflix_directors = netflix_titles.query('director != "No director"').set_index('title')['director'].str.split(',', expand = True).stack().reset_index(level = 1, drop = True)
plt.figure(figsize = (14,7))
sns.countplot(y = netflix_directors, order = netflix_directors.value_counts().index[:20], palette='bright')
plt.title('Топ 20 режиссеров по количеству контента')
plt.xlabel('Titles')
plt.ylabel('Directors')
plt.show()
````

````python
'''Наибольшее количество отснятых фильмов и тв шоу приходятся на мексиканских режиссеров''';
````

````python
'''
Следующим шагом будет анализ по самым популярным и не популярным жанрам фильмов и тв шоу.
''';

netflix_genres = netflix_titles.set_index('title')['listed_in'].str.split(',', expand = True).stack().reset_index(level = 1, drop = True)

fig, axs = plt.subplots(nrows = 2, ncols = 1, figsize = (10,10))

sns.countplot(y = netflix_genres, order = netflix_genres.value_counts().index[:20], ax = axs[0])
sns.countplot(y = netflix_genres, order = reversed(netflix_genres.value_counts().index[-20:]), ax = axs[1])
axs[0].set_title('Топ 20 популярных жанров на Netflix')
axs[1].set_title('Топ 20 не популярных жанров на Netflix')
axs[0].set_xlabel('Titles')
axs[0].set_ylabel('Genres')
axs[1].set_xlabel('Titles')
axs[1].set_ylabel('Genres')
plt.tight_layout()
plt.show()
````

````python
'''
Наиболее популярными жанрами явялются международные фильмы, драмы и комедии.
Наиболее не популярными жанрами по данным Netflix считаются Телевизионная научная фантастика и фэнтези, спортивные фильмы и 
тв шоу на испанском языке.
''';
````

````python
netflix_duration_filter = netflix_titles[netflix_titles['duration'].str.contains("Season")]
netflix_duration = netflix_duration_filter.set_index('title')['duration']

sns.countplot(y = netflix_duration, order = netflix_duration.value_counts().index[:20]);
````

````python
netflix_rating_unique = pd.DataFrame({'rating': netflix_titles['rating'].unique()})

count_movies = netflix_movies.groupby('rating')['title'].count().reset_index()
count_tvshows = netflix_tvshows.groupby('rating')['title'].count().reset_index()
count_tvshows = pd.merge(netflix_rating_unique, count_tvshows, on = 'rating', how='left').fillna(0)
count_tvshows = count_tvshows.set_index('rating').loc[count_movies['rating']].reset_index()

plt.figure(figsize=(13,7))
plt.title('Количество контента в разбивке по рейтингу (фильмы против ТВ шоу)')
plt.bar(count_movies['rating'], count_movies['title'], color = '#ff746a')
plt.bar(count_movies['rating'], count_tvshows['title'], bottom = count_movies['title'], color = '#ab95d4')
plt.legend(['Movies', 'TV Show'])
plt.xlabel('ratings')
plt.ylabel('titles')
plt.show()
````

````python
'''
Больше всего контента приходится на рейтинг "TV-MA". Данный рейтинг предназначен для просмотра людям старше 18 лет. 
Следующим не менее популярным рейтингом выступает "TV-14". Этот рейтинг предназначен на аудиторию старше 16 лет. 
''';
````

````python
'''
И на последок, сделаем анализ по самым популярным актёрам. Поделим их на фильмы и ТВ шоу.
''';

netflix_actor_tvshows = netflix_tvshows.query('cast != "No cast"').set_index('title')['cast'].str.split(',', expand = True).stack().reset_index(level = 1, drop = True).str.strip()
netflix_actor_movies = netflix_movies.query('cast != "No cast"').set_index('title')['cast'].str.split(',', expand = True).stack().reset_index(level = 1, drop = True).str.strip()

fig, axs = plt.subplots(nrows = 2, ncols = 1, figsize = (10,10))



sns.countplot(y = netflix_actor_tvshows, order = netflix_actor_tvshows.value_counts().index[:20], palette = 'bright', ax = axs[0])
sns.countplot(y = netflix_actor_movies, order = netflix_actor_movies.value_counts().index[:20], palette = 'bright', ax = axs[1])
axs[0].set_title('Топ-20 актеров по количеству ТВ шоу')
axs[1].set_title('Топ-20 актеров по количеству фильмов')
axs[0].set_xlabel('Titles')
axs[0].set_ylabel('Actors')
axs[1].set_xlabel('Titles')
axs[1].set_ylabel('Actors')


plt.tight_layout()
plt.show()


plt.show()
````
